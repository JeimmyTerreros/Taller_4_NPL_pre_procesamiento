```python
import re
import pandas as pd
import numpy as np
from collections import Counter

from nltk.corpus import stopwords
stopwords_sp = stopwords.words('spanish')

from sklearn.feature_extraction.text import CountVectorizer
```

```python
import nltk
nltk.download('stopwords') 
```

```python
archivo = pd.read_csv('C:/Users/yayat/OneDrive/Documentos/Jeimmy/Universidad/NPL/Taller 4/dialogos.csv')
df = pd.DataFrame(archivo)
df.head(1000)
```

```python
archivo1 = df.to_string()
print(archivo1)
```

```python
procesamiento = archivo1.lower() 
procesamiento = re.sub(r"[\W\d]", " ", procesamiento) 
procesamiento = procesamiento.split()
procesamiento = [palabra for palabra in procesamiento if palabra not in stopwords_sp]
procesamiento

```

```python
df1 = pd.DataFrame(procesamiento,columns=['pre_posesamiento'])
df1.head(1000)
```

```python
df = pd.DataFrame(archivo,columns=['Locución']) 
df['pre_prosesamiento'] = df1
df.head(1000)
```

```python
diccionario_frecuencias = {}
for palabra in procesamiento:
    if palabra in diccionario_frecuencias:
        diccionario_frecuencias[palabra]+=1
    else:
        diccionario_frecuencias[palabra]=1
    
```

```python
for palabra in diccionario_frecuencias:
    frecuencia = diccionario_frecuencias[palabra]
    print(f"'{palabra}'{frecuencia}")
```
